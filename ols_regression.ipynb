```python
import os
import numpy as np
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.iolib.summary2 import summary_col
```

```python
# Set directory path to current directory
dir_path = os.path.abspath('')

# Getting the data for anlaysis
dataset = os.path.join(dir_path, 'results_v1.xlsx')

# Turning the data into a dataframe
df = pd.read_excel(dataset, index_col=0, sheet_name="full_data")
```

```python
df.index = pd.Series(df.index).fillna(method='ffill')
```

```python
# Iterate through the dataframe and perform regression on each country

for name, group in df.groupby('country'):
    # print(name)
    y=group['elec_rate']

    # Model 1: The multiple regression model describes the response as a weighted sum of the predictors:(Elec = beta_0 + beta_1*ED + beta_2*EG,nr ~)
    X0 = group[['Energy distribution',
                'Energy generation, non-renewable sources',
                'Energy Policy',
                'Energy generation, renewable sources'                  
                ]]
    
    X0 = sm.add_constant(X0)
    
    # Model 2
    
    x0 = group['Energy distribution']
    x1 = group['Energy generation, non-renewable sources']
    x2 = group['Energy Policy']
    x3 = group['Energy generation, renewable sources']
    
    X1 = x0 + x1 + x2 + x3
    
    X1 = sm.add_constant(X1)
    
    reg0 = sm.OLS(y, X0).fit()
    reg1 = sm.OLS(y, X1).fit()
    
    reg0.summary()
    
    # results = summary_col([reg0],stars=True,float_format='%0.2f',
    #                    model_names=['Model 1'],
    #                    info_dict={'N':lambda x: "{0:d}".format(int(x.nobs)),
    #                               'R2':lambda x: "{:.2f}".format(x.rsquared)})

    # print(results) 
    
    # print('Country: {}  \nR^2 value for model 1 is: {:.2f},  \nfor model 2 is: {:.2f}'.format(name, reg0.rsquared, reg1.rsquared))
```

```python
results = summary_col([reg0],stars=True,float_format='%0.2f',
                       model_names=['Model 1'],
                       info_dict={'N':lambda x: "{0:d}".format(int(x.nobs)),
                                  'R2':lambda x: "{:.2f}".format(x.rsquared)})

print(results)
```
