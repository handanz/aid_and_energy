```python
import os
import numpy as np
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set directory path to current directory
dir_path = os.path.abspath('../')
dat = os.path.join(dir_path, 'control_variables.xlsx')
```

```python
df = pd.read_excel(dat, sheet_name="data")
df = pd.DataFrame(df)
```

```python
# Making the heat map for electricity rate in SSA

pivot_df = pd.pivot_table(df, index=['country'], columns='year', values=['elec_rate'])

sorted_df = pivot_df.reindex(pivot_df['elec_rate'].sort_values(by=2019).index)

sns.set(rc={'figure.figsize':(11.7,8.27)})
ax=sns.heatmap(sorted_df, xticklabels=1, cmap="YlGnBu")
```

```python
# Making the heat map for electricity rate in SSA

pivot_df = pd.pivot_table(df, index=['country'], columns='year', values=['gdp'])

sorted_df = pivot_df.reindex(pivot_df['gdp'].sort_values(by=2019).index)

ax2=sns.heatmap(sorted_df, xticklabels=1, cmap="OrRd")
```

```python
df['aid mil'] = df['energy aid']/1000000
aid_df = df.pivot('country', 'year', 'aid mil')
ax1=sns.heatmap(aid_df, cmap='PuBu')
```

```python
govern_df = pd.pivot_table(df, index=['country'], columns='year', values=['governance'])

govern_df = govern_df.reindex(govern_df['governance'].sort_values(by=2019).index)

sns.set(rc={'figure.figsize':(11.7,8.27)})

ax_govern=sns.heatmap(govern_df, cmap='PuBu')
```

```python
# comparing between electric rate and gdp
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

dfelec = df.pivot('country', 'year', 'elec_rate')
dfgdp = df.pivot('country', 'year', 'gdp')

fig, (ax_elec, ax_gdp) = plt.subplots(ncols=2)
fig.subplots_adjust(wspace=0.01)

sns.heatmap(dfelec, cmap="OrRd", ax=ax_elec, cbar=False)
sns.heatmap(dfgdp, cmap="OrRd", ax=ax_gdp, cbar=False)

ax_gdp.yaxis.tick_right()

fig.subplots_adjust(wspace=0.001)
plt.show()
```
